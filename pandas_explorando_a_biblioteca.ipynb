#TRATANDO E FILTRANDO OS DADOS


Dados Nulos

```python
df.isnull()
```

```python
df.isnull().sum() #somando todos os valores nulos
```

```python
df=df.fillna(0) # transformando valores nulos em 0
```

```python
df.isnull().sum()
```

```python
df.query('Valor == 0 or Condominio == 0').index # pegando o index de cada valor 0 na base de dados
```

```python
registro_remover = df.query('Valor == 0 or Condominio == 0').index
```

```python
df = df.drop(registro_remover) #apagando todos o valores zero na base de dados
```

```python
df.query('Valor == 0 or Condominio == 0')
```

```python
df.head(20)
```

```python
df.drop(columns='Tipo', inplace=True) # removendo coluna tipo, pois já sabemos que o dataframe é apenas apartamentos
```

```python
df.head()
```

```python
df_quartos=df['Quartos'] == 1
df[df_quartos]
```

```python
df_valor=df['Valor'] < 1200
df[df_valor]
```

```python
selecao_final = (df_valor) & (df_quartos)
```

```python
filtro_um = df[selecao_final]
filtro_um
```

```python
filtro_dois = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)
```

```python
df[filtro_dois]
```

```python
df.to_csv('Apartamentos.csv', index=False)
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
```

```python
dados = pd.read_csv(url, sep = ';')
```

```python
dados.head(5)
```

```python
dados['valor_por_mes'] = dados['Valor'] + dados['Condominio']
```

```python
dados.head()
```

```python
dados['valor_por_ano'] = (dados['valor_por_mes'] * 12) + dados['IPTU']
```

```python
dados.head()
```

```python
dados['descricao'] = 'Tipo de locação: ' + dados['Tipo'] + ', Bairro: ' + dados['Bairro'] + ', Quantia de quartos: ' + dados['Quartos'
].astype(str) + ', Vaga de garagem: ' + dados['Vagas'].astype(str)
```

```python
dados['descricao']
```

```python
dados['Possui Suite'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Não')
```
